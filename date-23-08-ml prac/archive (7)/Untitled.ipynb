```python
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=['highest_monthly_earnings']),
                                                 df['highest_monthly_earnings'],
                                                 test_size=0.2,
                                                random_state=42)
```

```python
X_train.head()
```

```python
from sklearn.compose import ColumnTransformer
# imputation transformer
trf1 = ColumnTransformer([
    ('impute_country',SimpleImputer(strategy='most_frequent'),[3]),
    ('impute_chanel',SimpleImputer(strategy='most_frequent'),[4])
],remainder='passthrough')
```

```python
from sklearn.preprocessing import OneHotEncoder

# one hot encoding
trf2 = ColumnTransformer([
    ('ohe_sex_embarked',OneHotEncoder(sparse=False,handle_unknown='ignore'),[1,6])
],remainder='passthrough')
```

```python
from sklearn.preprocessing import MinMaxScaler

trf3 = ColumnTransformer([
    ('scale',MinMaxScaler(),slice(0,10))
])
```

```python
from sklearn.tree import DecisionTreeClassifier
trf4 = DecisionTreeClassifier()
```

```python
from sklearn.pipeline import Pipeline,make_pipeline

pipe = Pipeline([
                ('trf1',trf1),
                ('trf2',trf2),
                ('trf3',trf3),
                ('trf4',trf4),
            ])
```

```python
pipe = make_pipeline(trf1,trf2,trf3,trf4)
#pipe.fit(X_train,y_train)
```

```python
from sklearn import set_config
set_config()
```

```python
y_pred=pipe.predict(X_test)
```
