```python
def ctgry(price):
    if price>car_data.Price.median():
        return("expensive")
    else:
        return("cheap")
```

```python
car_data["category_by_lambda"]=car_data.Price.apply(lambda p:"expensive" if p>car_data.Price.median() else "cheap")
```
